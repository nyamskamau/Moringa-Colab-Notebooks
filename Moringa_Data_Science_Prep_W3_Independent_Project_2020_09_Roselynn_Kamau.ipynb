## **DATA CLEANING**

**The first step of this process was renaming columns in Dataset1 and three then joining them 
function.**


```
df1 = pd.read_csv('Telcom_dataset.csv')
df1.rename(columns={'PRODUTC' :'PRODUCT' , 'DW_A_NUMBER_INT': 'DW_A_NUMBER', 'DW_B_NUMBER_INT': 'DW_B_NUMBER','DATETIME' : 'DATE_TIME'}, inplace=True)
df1.head(2)
```

```
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.rename(columns={'DW_A_NUMBER_INT': 'DW_A_NUMBER', 'DW_B_NUMBER_INT': 'DW_B_NUMBER','CELLID' : 'CELL_ID' , 'SIET_ID' : 'SITE_ID'}, inplace=True)
df3.head(2)
```

```
df4 = pd.concat([df1 ,df2,df3])
df4
```

**Then the COUNTRY_A , COUNTRY_B and CELL_ON_SITE columns were dropped from the dataset.**


```
df4 = df4.drop(['COUNTRY_A',"COUNTRY_B",'CELL_ON_SITE'] , axis = 1)
df4.head()
```

**The subsequent table was then merged with the Cells_Geo table , null values and duplicates values dropped**

```
df = df4.merge(df5 , left_index=True , right_index=True)
df.head()
```

```
df1 = df['DATE_TIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(' '))]))
df1.columns = ['TIME', 'DATE']
df1
df7 = df1.merge(df, left_index=True, right_index=True)
df7
#droping DATE_TIME column
df8 = df7.drop('DATE_TIME', axis=1)
df8.head()
```

```
df = df.dropna()
df.head(2)
```

```
df = df.drop_duplicates()
df.head()
```
